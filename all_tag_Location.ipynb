# Location to txt file

```python
import numpy as np
import pandas as pd
import os
import datetime
from tqdm import tqdm
import glob 
'''
Don't forget to change the data_dir!!
'''

data_dir = r'E:\Instagram\#태풍\location'
os.chdir(data_dir)
l_files = glob.glob('*_location.txt')
post_date = [datetime.datetime.strptime(i[:-18],'%Y-%m-%d_%H-%M-%S') for i in l_files]


x=[]
for i in tqdm(l_files):
    f = open(data_dir+'/'+i,'r', encoding='UTF8') 
    x.append(f.readlines())
    
```

```python
# Remove \n
x = [[i.strip('\n') for i in j] for j in x]
loc = []
g_map = []
for i in tqdm(range(len(x))):
    loc.append(x[i][0])
    g_map.append(x[i][1])
```

```python
x
```

## Creating a dataframe

```python
loc_df = pd.DataFrame({'Post_created_time':post_date,
                      'location':loc,
                      'GoogleMap':g_map})
```

```python
coord = [x[31:x.find("&")] for x in g_map]
lat = [x[0:x.find(",")] for x in coord]
lng = [x[(x.find(",")+1):] for x in coord]

loc_df['Latitude'] =lat
loc_df['Longitude'] = lng
```

```python
loc_df.columns
```

```python
loc_df = loc_df.sort_values('Post_created_time')
loc_df = loc_df.loc[loc_df['Post_created_time'] > '2019-09-01']
```

```python
loc_df = loc_df[loc_df.Latitude != 'None']
```
